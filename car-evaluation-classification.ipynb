```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
car_evaluation = fetch_ucirepo(id=19) 
  
# data (as pandas dataframes) 
X = car_evaluation.data.features 
y = car_evaluation.data.targets 
  
# metadata 
print(car_evaluation.metadata) 
  
# variable information 
print(car_evaluation.variables) 
```

```python
from sklearn.preprocessing import LabelEncoder 
encoder = LabelEncoder()
for i in ['buying', 'maint','lug_boot','safety']:
    car[i] = encoder.fit_transform(car[[i]])
    
```

```python
y['class'] = encoder.fit_transform(y[['class']])
```

```python
car.tail(30)
```

```python
 
car['doors'] = car['doors'].replace('5more', 5).astype(int)
car['persons'] = car['persons'].replace('more', 5).astype(int)
```

```python
car.tail()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

xtrain,xtest,ytrain,ytest = train_test_split(car,y,test_size =0.23, random_state = 42)
```

```python
models = {'lr':LogisticRegression(),
'gbc':GradientBoostingClassifier(),'ada':
AdaBoostClassifier(),
'svm':SVC()
,'knn' : KNeighborsClassifier()
,'dt':DecisionTreeClassifier(),
'xgb':xgb.XGBClassifier(),
         'rfc': RandomForestClassifier()}
```

```python
for keys , values in models.items():
    print(keys)
    values.fit(xtrain,ytrain)
    print(f'Training accuracy : {values.score(xtrain,ytrain)}  Testing Accuracy : {values.score(xtest,ytest)}')
    print("-------")
```

```python
#after scaling 
scaler = StandardScaler()
for i in car.columns:
    car[i] = scaler.fit_transform(car[[i]])
```

```python
xtrains,xtests,ytrains,ytests = train_test_split(car,y , test_size=0.23, random_state = 42)
```

```python
for keys , values in models.items():
    print(keys)
    values.fit(xtrains,ytrains)
    print(f'Training accuracy : {values.score(xtrains,ytrains)}  Testing Accuracy : {values.score(xtests,ytests)}')
    print("-------")
```
