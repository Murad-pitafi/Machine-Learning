```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')
# Load and preprocess the MNIST dataset
(x_train, y_train), (_, _) = mnist.load_data()
x_train = x_train / 255.0
x_train = x_train.reshape((-1, 28, 28, 1))
y_train = to_categorical(y_train, 10)

```

```python
import os

 
filepath = 'supervised_weights.weights.h5'

 
exists = os.path.exists(filepath)

 
print(exists)
```

```python
 
supervised_model = Sequential([
    Flatten(input_shape=(28, 28, 1)),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='softmax')
])
supervised_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
supervised_model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

 
supervised_model.save_weights('/kaggle/working/supervised.weights.h5')
```

```python
from tensorflow.keras.layers import Reshape

 
autoencoder = Sequential([
    Flatten(input_shape=(28, 28, 1)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(784, activation='sigmoid'),
    Reshape((28, 28, 1))
])
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.fit(x_train, x_train, epochs=10, batch_size=32, validation_split=0.2)

 
autoencoder.save_weights('/kaggle/working/unsupervised.weights.h5')
```

```python
supervised_model.load_weights('/kaggle/working/supervised.weights.h5')
 
supervised_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
supervised_model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```
