```python
email = pd.read_csv('/kaggle/input/email-spam-classification-dataset/combined_data.csv')
email.head(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
x = email['text']
y = email['label']

xtrain, xtest, ytrain , ytest = train_test_split(x,y , test_size = 0.23)


from sklearn.feature_extraction.text import TfidfVectorizer


from sklearn.feature_extraction.text import TfidfVectorizer


from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(xtrain)
X_test_features = feature_extraction.transform(xtest)

ytrain = ytrain.astype('int')
ytest = ytest.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression 
logistic_reg = LogisticRegression()
logistic_reg.fit(X_train_features, ytrain)

print(logistic_reg.score(X_train_features, ytrain))
print(logistic_reg.score(X_test_features, ytest))
```

Let's try on real incoming email 

```python
samples = ["Hello! We are giving discounts on this new offer.",
           "You have been nominated for the Best Award."]

sample_features = feature_extraction.transform(samples)


results = logistic_reg.predict(sample_features)

print("Predictions:", results)
```
